```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/alexeygrigorev/datasets/master/car_fuel_efficiency.csv')
```

```python
num_rows = df.shape[0]
print(num_rows)
```

```python
unique_fuel = df['fuel_type'].unique()
print(len(unique_fuel))
```

```python
 print(df.isnull().sum())
```

```python
max_value = df['fuel_efficiency_mpg'].max()
print(max_value)
```

```python
unique_origin = df['origin'].unique()
print(unique_origin)
```

```python
filtered_df = df[df['origin'] == 'Asia']
max_value = filtered_df['fuel_efficiency_mpg'].max()
print(max_value)
```

```python
init_median = df['horsepower'].median()
print(init_median)
```

```python
mode = df['horsepower'].mode()
print(mode)
```

```python
df_copy = df.copy()
df_copy['horsepower'] = df_copy['horsepower'].fillna(mode)
final_median = df_copy['horsepower'].median()
print(final_median)
```

```python
delta_median = final_median - init_median
print(delta_median)
```

```python
asia_df = df[df['origin'] == 'Asia'][['vehicle_weight','model_year']].head(7)
print(asia_df)
```

```python
X = asia_df.to_numpy()
print(X)
```

```python
X_transpose = X.T
print(X_transpose)
```

```python
XTX = X_transpose @ X
print(XTX)
```

```python
import numpy as np
```

```python
XTX_invert = np.linalg.inv(XTX)
print(XTX_invert)
```
